```python
class Config:
    host = "localhost"
    port = 8080
    debug = False
    @classmethod
    def from_dict(cls, d):
        cls.host = d.get("host")
        cls.port = d.get("port")
        cls.debug = d.get("debug")
        # TODO: Update class variables from dictionary keys
        
    @classmethod
    def display(cls):
        # TODO: Print host, port, debug
        print(Config.host)
        print(Config.port)
        print(Config.debug)

# Test:
Config.display() # localhost:8080, debug=False
Config.from_dict({"host": "0.0.0.0", "port": 5000, "debug": True})
Config.display() # 0.0.0.0:5000, debug=True
```
